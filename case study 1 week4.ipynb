```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read the dataset to the python environment.

```python
car_data=pd.read_csv("cars_data.csv")
```

# plot a graph of male vs female buyers participated in the sales

```python
plt.figure(figsize=(10,8))
sns.countplot(x=car_data['Buyer Gender'],data=car_data)
plt.xlabel('gender',fontsize=18)
plt.ylabel('Frequency',fontsize=18)
plt.title('Plot of Buyer Gender vs Frequency',fontsize=20,weight='bold')
plt.grid(linestyle= '--')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(x='Buyer Gender',y='Sale Price',data=car_data)
plt.title('plot of sale price vs Gender')
plt.grid(linestyle= '--')
```

# find top 5 cars based on their sale price

```python
data1=car_data.sort_values(by=['Sale Price'],ascending=False)[['Sale Price','Make','Model']].head(5)
```

```python
data1
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data1['Make'],hue=data1['Sale Price'],data=data1)
plt.xlabel('Make')
plt.show()
```

# Find the least 5 cars based on their Resell price

```python
data2=car_data.sort_values(by=['Resell Price'],ascending=False)[['Resell Price','Make','Model']].tail(5)
```

```python
data2
```

```python
plt.figure(figsize=(8,6))
sns.countplot(data2['Make'],hue=data2['Resell Price'],data=data2)
plt.xlabel('Make')
plt.show()
```
